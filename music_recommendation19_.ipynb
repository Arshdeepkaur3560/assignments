```python
# STEP 1: Upload the CSV file
from google.colab import files
uploaded = files.upload()

# STEP 2: Import libraries
import pandas as pd

# STEP 3: Load the CSV (replace the filename if needed)
df = pd.read_csv("music_recommendation_dataset_200 (1).csv")

# STEP 4: Preview the data
print("First 5 rows:")
print(df.head())

# STEP 5: Summary statistics
print("\nSummary statistics:")
print(df.describe(include='all'))

# STEP 6: Check for missing values
print("\nMissing values:")
print(df.isnull().sum())

# Optional: Show number of unique genres
print("\nUnique genres:")
print(df['genre'].value_counts())

# Optional: Plot genre distribution
import matplotlib.pyplot as plt

df['genre'].value_counts().plot(kind='bar', title='Genre Distribution', figsize=(10,5))
plt.xlabel("Genre")
plt.ylabel("Number of Songs")
plt.show()
```
